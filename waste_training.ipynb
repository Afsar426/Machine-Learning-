```python
df = df.drop(columns=[
    "Day",
    "T Value",
    "Score"
])

print(df.head())
```

```python
def assign_waste_type(cn):
    if cn < 20:
        return "Vegetable"
    elif cn <= 30:
        return "Mixed"
    else:
        return "Dry Leaves"

df["Waste_Type"] = df["C/N Ratio"].apply(assign_waste_type)

print(df[["C/N Ratio", "Waste_Type"]])
```

```python
def assign_aeration(mc):
    if mc < 40:
        return "High"
    elif mc <= 60:
        return "Medium"
    else:
        return "Low"

df["Aeration"] = df["MC(%)"].apply(assign_aeration)

print(df[["MC(%)","Aeration"]])
```

```python
def assign_smell_level(ammonia):
    if ammonia < 50:
        return "Low"
    elif ammonia <= 150:
        return "Medium"
    else:
        return "High"

df["Smell_Level"] = df["Ammonia(mg/kg)"].apply(assign_smell_level)

print(df[["Ammonia(mg/kg)","Smell_Level"]])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_waste = LabelEncoder()
le_aeration = LabelEncoder()
le_smell = LabelEncoder()

df["Waste_Type"] = le_waste.fit_transform(df["Waste_Type"])
df["Aeration"] = le_aeration.fit_transform(df["Aeration"])
df["Smell_Level"] = le_smell.fit_transform(df["Smell_Level"])
print(df.head())
```

```python
def compost_status(gi):
    if gi >= 80:
        return "Ready"
    else:
        return "Not Ready"

df["Compost_Status"] = df["GI(%)"].apply(compost_status)
```

```python
df = df.drop(columns=["GI(%)"])
```

# Compost Modal

```python

X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]

print("X columns:", X.columns)
print("y name:", y.name)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)
```

```python
X_train = X_train.values
X_test = X_test.values
```

```python
model = RandomForestClassifier(n_estimators=50, random_state=1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("acc:", accuracy_score(y_test, y_pred))
```
